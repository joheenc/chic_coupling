```python
from astropy import units as u
from poliastro.bodies import Earth, Sun
from poliastro.twobody import Orbit
a = 10000. * u.km #semimajor axis
ecc = 0. * u.one #eccentricity
inc = 0. * u.deg #inclination
raan = 0. * u.deg #RA of ascending node
argp = 0. * u.deg #argument of periastron
nu = 90. * u.deg #true anomaly (start)
orb = Orbit.from_classical(Earth, a, ecc, inc, raan, argp, nu)
orb.plot()
```

```python
from astropy import time
from datetime import datetime
epoch = time.Time(datetime.now())
from poliastro.ephem import Ephem
earth = Ephem.from_body(Earth, epoch.tdb)
from astropy.coordinates import solar_system_ephemeris
solar_system_ephemeris.set("jpl")
orb = Orbit.from_ephem(Sun, earth, epoch)
orb.plot()
```

```python
orb=orb.propagate(100*u.day)
orb.plot()
```
